```python
#Write a Python program to extract all the URLs from the webpage python.org that are nested within <li> tags from
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('https://www.python.org/?','html.parser')
start=requests.get(url)
s1=start.text
soup=BeautifulSoup(s1,'lxml')
list1=[]
for i in soup.find_all('li'):
    b=i.find('a')
    list1.append(b.attrs['href'])
print(list1)
```

```python
#Write a Python program to retrieve children of the html tag from a given web page.
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('https://www.python.org/?','html.parser')
start=requests.get(url)
soup=BeautifulSoup(start.text,'lxml')
root=soup.html
root_child=[e.name for e in root.children if e.name is not None]
print(root_child)
```

```python
#Write a Python program to retrieve all descendants of the body tag from a given web page.
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('https://www.python.org/?','html.parser')
start=requests.get(url)
soup=BeautifulSoup(start.text,'lxml')
root=soup.html
root_child=[e.name for e in root.descendants if e.name is not None]
print(set(root_child))
```

```python
#Write a Python program to print content of elements that contain a specified string of a given web page. 
from bs4 import BeautifulSoup
import requests
import re
url=BeautifulSoup('https://www.python.org/?','html.parser')
start=requests.get(url)
soup=BeautifulSoup(start.text,'lxml')
for i in soup.find_all(string=re.compile('python')):
    print(i.split())
```
