```python
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('https://www.worldometers.info/coronavirus/','html.parser')
start=requests.get(url)
s1=start.text
soup=BeautifulSoup(s1,'lxml')
tags=soup.table

tag=tags.find_all('tr')
list1=[]
for i in range(len(tag)):
    x=tag[i].get_text()
    list1.append(x)
del list1[1:7]
del list1[1:3]
list1=list1[:216]
final=[]
for j in list1:
 y=j.split('\n')
 final.append(y)
final
```

```python
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('http://www.nepalstock.com/floorsheet','html.parser')
response=requests.get(url)
data=response.text
soup=BeautifulSoup(data,'lxml')
#soup
tag=soup.table
tags=tag.find_all('tr')
#tags
list1=[]
for i in range(len(tags)):
    x=tags[i].get_text()
    list1.append(x)
#list1
list1=list1[1:-3]
list2=[]
for i in list1:
    y=i.split('\n')[1:-1]
    list2.append(y)
list2
```

```python
import csv
n = len(list2)
with open('nepal.csv','a') as csvFile:
  writer = csv.writer(csvFile)
  for i in range(n):
    writer.writerow(list2[i])
```

```python
#for multiple
from bs4 import BeautifulSoup
import requests
n=int(input("how many page you want?"))
final=[]
for i in range(1,n+1):
    url=BeautifulSoup(f"http://www.nepalstock.com/main/floorsheet/index/{i}/","html.parser")
    response=requests.get(url)
    data=response.text
    soup=BeautifulSoup(data,'lxml')
    #soup
    tag=soup.table
    tags=tag.find_all('tr')
    #tags
    list1=[]
    for i in range(len(tags)):
        x=tags[i].get_text()
        list1.append(x)
    #list1
    list1=list1[1:-3]
    list2=[]
    for i in list1:
        y=i.split('\n')[1:-1]
        list2.append(y)
    final=final+list2
final
```

```python
import csv
n = len(final)
with open('final.csv','a') as csvFile:
  writer = csv.writer(csvFile)
  for i in range(n):
    writer.writerow(final[i])
```

```python
import pandas as pd
df = pd.read_csv('final.csv')
df
```
