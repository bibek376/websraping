```python
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup("http://seismonepal.gov.np/earthquakes/2019","html.parser")
start=requests.get(url)
s1=start.text
soup=BeautifulSoup(s1,"lxml")
tags=soup.table
tag=tags.find_all("tr")
list1=[]
for i in range(len(tag)):
    x=tag[i].get_text()
    list1.append(x)
list2=[]
for i in list1:
    y=i.split("\n")[1:-2]
    list2.append(y)
for j in range(len(list2)-1):
    del list2[j+1][0]
print(len(list2))
for k in range(len(list2)-1):
    del list2[k+1][1:3]
for l in range(len(list2)-1):
    del list2[l+1][2]
list2
```

```python
import csv
import pandas as pd
with open("carlos.csv","w+") as f:
    x=csv.writer(f)
    x.writerows(list2)
df=pd.read_csv("carlos.csv")
df
```
